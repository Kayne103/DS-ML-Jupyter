```python
import pandas as pd
carSales = pd.read_csv("car-sales.csv") 
```

```python
carSales
```

```python
carSales["Price"] = carSales["Price"].str.replace('[\$\,\.]','')
carSales
```

```python
carSales["Price"] = carSales["Price"].str[:-2]
carSales
```

```python
#Add a date column
carSales["Date"] = pd.date_range("1/1/2020",periods=len(carSales))
carSales
```

```python
#Add a total column
carSales["Total_Sales"] = carSales["Price"].astype(int).cumsum()
carSales
```

```python
carSales.plot(x='Date',y = 'Total_Sales')
```

```python
carSales["Price"] = carSales["Price"].astype(int)
carSales.plot(x="Odometer (KM)",y="Price",kind="scatter")
```

```python
import pandas as pd
heartDisease = pd.read_csv("heart-disease.csv")
heartDisease
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
heartDisease["age"].plot.hist(bins=50)
```

```python
#how to select some set of data.
heartDisease.head(10)
```

```python
heartDisease.tail(10)
```

```python
#select data that is over fifty
AgeOverFifty = heartDisease[heartDisease["age"]>50]
AgeOverFifty
```

```python
AgeOverFifty.head()
```

```python
AgeOverFifty.plot(x="age",y="chol",kind="scatter")
```

```python
AgeOverFifty.plot.scatter(x="age",y="chol",c="target",figsize=(10,6))
```

```python
#design a more complicated and full plot
#create the plot
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter =ax.scatter(AgeOverFifty["age"],AgeOverFifty["chol"],c=AgeOverFifty["target"])
#customize the plot
ax.set(title="Government Hospital Heart Disease and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title ="Target")
ax.axhline(AgeOverFifty["chol"].mean(),linestyle="--",label= "Average Cholestrol Level")
```
