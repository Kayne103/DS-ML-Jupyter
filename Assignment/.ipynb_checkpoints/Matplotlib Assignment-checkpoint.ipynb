```python
#Preparing data for a bar plot
myFood = {"hotdog":15.00,"chips":7.00,"fatcake":1.50}
myFood
```

```python
# plot a bar plot
fig,ax = plt.subplots()
ax.bar(list(myFood.keys()),list(myFood.values()))
# style the bar plot
ax.set(title="My Food",xlabel="Price(BWP)",ylabel="Food Name")
```

```python
# plot a horizontal bar plot
fig,ax = plt.subplots()
ax.barh(list(myFood.keys()),list(myFood.values()))
# style the bar plot
ax.set(title="My Food",xlabel="Price(BWP)",ylabel="Food Name")
```

```python
# A 2 by 2 empty subplot
fig, ax = plt.subplots(2,2)
```

```python
# A 2 by 2 subplot
fig, ax = plt.subplots(2,2,figsize=(12,10))#changed figsize so that labels can be visible

# Plot X versus X/2 on the top left subplot
ax[0][0].plot(randomNumbers,randomNumbers/2)
ax[0][0].set(title="X versus X/2",xlabel="X/2",ylabel="X")

# Plot a scatter plot on the top right subplot
X = np.random.randn(10)# generate 10 random numbers
ax[0][1].scatter(X,X)
ax[0][1].set(title="Scatter plot",xlabel="X",ylabel="X")

# Plot a bar plot on the bottom left of the subplot, using myFood dictionary
ax[1][0].bar(list(myFood.keys()),list(myFood.values()))
ax[1][0].set(title="Bar graph",xlabel="Price(BWP)",ylabel="Food Name")

# Plot a histogram of 1000 random normally distributed numbers
X = np.random.normal(size = 1000)
ax[1][1].hist(X)
ax[1][1].set(title="Histogram")

#save the fig (optional)
fig.savefig("2by2_subplot.png")
```

```python
import pandas as pd
```

```python
#read csv
carSales = pd.read_csv("car-sales.csv")
# view the dataset
carSales
```

```python
#plot the price column
plt.plot(carSales["Price"])
plt.show
```

```python
# Removing the symbols from the price column
carSales["Price"] = carSales["Price"].str.replace('[\$\,\.]','')

# Remove the final two numbers from the price
carSales["Price"] = carSales["Price"].str[:-2]

# Convert Price from string to integer
carSales["Price"] = carSales["Price"].astype(int)
```

```python
carSales
```

```python
# Add a column called 'Total Sales' to car_sales which cumulatively adds
carSales["Total_Sales"] = carSales["Price"].astype(int).cumsum()
#Add a column called 'Sale Date' which lists a series of successive dates
carSales["Date"] = pd.date_range("1/1/2020",periods=len(carSales))
# View the dataset
carSales
```

```python
carSales.plot(x='Date',y = 'Total_Sales')
```

```python
# A scatter plot of Odometer against Price
carSales.plot.scatter(x="Odometer (KM)",y="Price")
```

```python
# Import heart_disease dataset
heartDisease = pd.read_csv("heart-disease.csv")
```

```python
heartDisease.head(10)
```

```python
heartDisease["age"].plot.hist(bins=50)
```

```python
heartDisease.plot.hist(subplots=True)
```

```python
heartDisease.plot.hist(subplots=True,figsize=(10,30))
```

```python
#select data that is over fifty
AgeOverFifty = heartDisease[heartDisease["age"]>50]

#create a plot
AgeOverFifty.plot(x="age",y="chol",kind="scatter")
```

```python
#design a more complicated and full plot
#create the plot
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter =ax.scatter(AgeOverFifty["age"],AgeOverFifty["chol"],c=AgeOverFifty["target"])
#customize the plot
ax.set(title="Government Hospital Heart Disease and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title ="Target")
ax.axhline(AgeOverFifty["chol"].mean(),linestyle="--",label= "Average Cholestrol Level")
```

```python
plt.style.use("fivethirtyeight")
fig, ax = plt.subplots(figsize=(10,6))
#plot the data
scatter =ax.scatter(AgeOverFifty["age"],AgeOverFifty["chol"],c=AgeOverFifty["target"],cmap="winter")
#customize the plot
ax.set(title="Government Hospital Heart Disease and Cholestrol Levels",xlabel="Age",ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title ="Target")
ax.axhline(AgeOverFifty["chol"].mean(),linestyle="--",label= "Average Cholestrol Level",color="red")
```

```python
#save the above fig
fig.savefig("ageOverFifty.png")
```
